```python
import mujoco
import torch
import numpy as np

# 1. Create a simple world: A Floor, a Falling Cube, and a Fixed Finger
xml = """
<mujoco>
    <worldbody>
        <light pos="0 0 3"/>
        <geom type="plane" size="1 1 0.1" rgba=".9 .9 .9 1"/>
        <body name="cube" pos="0 0 0.2">
            <joint type="free"/>
            <geom type="box" size="0.05 0.05 0.05" rgba="1 0 0 1"/>
        </body>
        <body name="finger" pos="0.08 0 0.1">
            <geom type="sphere" size="0.04" rgba="0 0 1 1"/>
        </body>
    </worldbody>
</mujoco>
"""

# 2. Initialize the Physics
model = mujoco.MjModel.from_xml_string(xml)
data = mujoco.MjData(model)

# 3. Run a few steps of physics
for _ in range(100):
    mujoco.mj_step(model, data)

# 4. Check for Contacts (The CAND Secret Sauce)
n_contacts = data.ncon
print(f"✅ SUCCESS: System is live!")
print(f"MuJoCo Version: {mujoco.__version__}")
print(f"PyTorch Version: {torch.__version__}")
print(f"Current Contacts Detected: {n_contacts}")

if n_contacts > 0:
    print("🤖 The robot 'feels' the object!")
else:
    print("⚪ No contact yet, physics is running smoothly.")
```
